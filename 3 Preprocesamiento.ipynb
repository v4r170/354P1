# 3. Preprocesamiento en Python

```python
import numpy as np
from scipy import stats

import pandas as pd 
df = pd.read_csv("caesarian1.txt")

pd.set_option('display.max_rows',None)
#df.head()
#print(df)
df
```

IMPUTACION El dataset tiene campos vacios (NAN), como los datos son cualitativos, se procede a calcular la moda para cada columna

```python
moda=stats.mode(df)
print(moda)
print(moda[0][0])
m=moda[0][0];
```

```python
df["edad"]=df["edad"].replace(np.nan,m[0])
df["numero_partos"]=df["numero_partos"].replace(np.nan,m[1])
df["tiempo_parto"]=df["tiempo_parto"].replace(np.nan,m[2])
df["presion_sanguinea"]=df["presion_sanguinea"].replace(np.nan,m[3])
df["problema_corazon"]=df["problema_corazon"].replace(np.nan,m[4])
df
```

DISCRETIZAR, Si 1, no 0, valores cuantitativos

```python
df["class"]=df["class"].replace("Si",1)
df["class"]=df["class"].replace("No",0)

df
```

CATEGORIZAR, Rural=R, Urbana=U, Valores Cualitativos

```python
df["presion_sanguinea"]=df["presion_sanguinea"].replace("Alta","A")
df["presion_sanguinea"]=df["presion_sanguinea"].replace("Normal","N")
df["presion_sanguinea"]=df["presion_sanguinea"].replace("Baja","B")

df
```
